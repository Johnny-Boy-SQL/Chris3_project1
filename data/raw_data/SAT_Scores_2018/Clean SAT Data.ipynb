```python
clean_data_SAT = data_SAT[["SCHOOL_DISTRCT_CD", "SCHOOL_DSTRCT_NM", "TEST_CMPNT_TYP_CD", "STATE_AVG_SCORE_VAL", "DSTRCT_AVG_SCORE_VAL"]]
clean_data_SAT.head(20)
```

```python
# cleaner_data_SAT = clean_data_SAT.loc[(clean_data_SAT["TEST_CMPNT_TYP_CD"] == "Combined Test Score") |  
#                                     (clean_data_SAT["TEST_CMPNT_TYP_CD"] == "Math Section Score - New") |
#                                     (clean_data_SAT["TEST_CMPNT_TYP_CD"] == "Evidence Based Reading and Writing - New"), :]

cleaner_data_SAT = clean_data_SAT.loc[(clean_data_SAT["TEST_CMPNT_TYP_CD"] == "Combined Test Score"), :]

cleaner_data_SAT

# cleaner_data_SAT = clean_data_SAT[clean_data_SAT["TEST_CMPNT_TYP_CD"].isin(["Math Section Score - New"])]

# cleaner_data_SAT
```

```python
# drug_gp_std = clean_comb_data.groupby(["Drug Regimen"])["Tumor Volume (mm3)"].std()
comb_score_std = cleaner_data_SAT.groupby(["TEST_CMPNT_TYP_CD"])["DSTRCT_AVG_SCORE_VAL"].std()
comb_score_std = int(comb_score_std)
comb_score_std
```

```python
comb_state_avg = cleaner_data_SAT["STATE_AVG_SCORE_VAL"].max()
comb_state_avg
```

```python
#bins = [0, (comb_state_avg - comb_score_std), (comb_state_avg + comb_score_std), 1600]
#group_labels = ["Below Average", "Average", "Average", "Above Average"]

#cleaner_data_SAT["Combined Score Bracket"] = pd.cut(cleaner_data_SAT[""])

# np.where(cleaner_data_SAT["DSTRCT_AVG_SCORE_VAL"] < (cleaner_data_SAT["STATE_AVG_SCORE_VAL"] - buffer), "Below Average")

# ***

conditions = [
    (cleaner_data_SAT["DSTRCT_AVG_SCORE_VAL"] < (cleaner_data_SAT["STATE_AVG_SCORE_VAL"] - comb_score_std)),
    (cleaner_data_SAT["DSTRCT_AVG_SCORE_VAL"] >= (cleaner_data_SAT["STATE_AVG_SCORE_VAL"] - comb_score_std)) & (cleaner_data_SAT["DSTRCT_AVG_SCORE_VAL"] <= (cleaner_data_SAT["STATE_AVG_SCORE_VAL"] + comb_score_std)),
    (cleaner_data_SAT["DSTRCT_AVG_SCORE_VAL"] > (cleaner_data_SAT["STATE_AVG_SCORE_VAL"] + comb_score_std))
             ]

values = ["Below Average", "Average", "Above Average"]

cleaner_data_SAT["Test Score Bracket"] = np.select(conditions, values)
cleaner_data_SAT
```

```python
#Drop NA records
cleaner_data_SAT = cleaner_data_SAT.dropna()
cleaner_data_SAT
```

```python
# df.rename(columns={"A": "a", "B": "c"})
cleaner_data_SAT = cleaner_data_SAT.rename(columns = 
                        {"SCHOOL_DISTRCT_CD": "School Dist Code",
                         "SCHOOL_DSTRCT_NM": "School Dist Name",
                         "TEST_CMPNT_TYP_CD": "Test Component",
                         "STATE_AVG_SCORE_VAL": "State Avg Score",
                         "DSTRCT_AVG_SCORE_VAL": "Dist Avg Score"
                        }) 
cleaner_data_SAT
```

```python
cleaner_data_SAT.to_csv("cleaner_data_SAT.csv", index = False)
```
