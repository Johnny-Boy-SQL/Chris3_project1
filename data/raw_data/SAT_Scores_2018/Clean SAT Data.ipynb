```python
comb_state_avg = 1052
buffer = 50

bins = [0, (comb_state_avg-buffer), (comb_state_avg+buffer), 1600]
group_labels = ["Below Average", "Average", "Average", "Above Average"]

cleaner_data_SAT["Combined Score Bracket"] = pd.cut(cleaner_data_SAT[""])

# np.where(cleaner_data_SAT["DSTRCT_AVG_SCORE_VAL"] < (cleaner_data_SAT["STATE_AVG_SCORE_VAL"] - buffer), "Below Average")

# ***

conditions = [
    (cleaner_data_SAT["DSTRCT_AVG_SCORE_VAL"] < (cleaner_data_SAT["STATE_AVG_SCORE_VAL"] - buffer)),
    (cleaner_data_SAT["DSTRCT_AVG_SCORE_VAL"] >= (cleaner_data_SAT["STATE_AVG_SCORE_VAL"] - buffer)) & (cleaner_data_SAT["DSTRCT_AVG_SCORE_VAL"] <= (cleaner_data_SAT["STATE_AVG_SCORE_VAL"] + buffer)),
    (cleaner_data_SAT["DSTRCT_AVG_SCORE_VAL"] > (cleaner_data_SAT["STATE_AVG_SCORE_VAL"] + buffer))
             ]

values = ["Below Average", "Average", "Above Average"]

cleaner_data_SAT["Bracket"] = np.select(conditions, values)
cleaner_data_SAT
```
