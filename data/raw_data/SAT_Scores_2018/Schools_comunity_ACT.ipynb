```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import linregress
```

```python
#files with data for analysis
# school_data = "raw_data/SAT_Scores_2018/cleaner_data_SAT.csv"
# School_Location = "raw_data/SAT_Scores_2018/free_reduced_lunch1718.csv"

school_data = "cleaner_data_SAT.csv"
School_Location = "free_reduced_lunch1718.csv"
```

```python
schools = pd.read_csv(school_data)
schools_type = pd.read_csv(School_Location)
```

```python
schools.head(5)
```

```python

schools_type = schools_type[['State Agency ID [District] 2017-18','Agency Name',
                             'Urban-centric Locale [District] 2017-18', 'Total Race/Ethnicity [District] 2017-18',
                            'Free and Reduced Lunch Students [Public School] 2017-18']]
schools_type
#All_school = pd.merge(schools, students, how="left", on=["school_name", "school_name"])
```

```python
schools_type = schools_type.rename(columns={'State Agency ID [District] 2017-18':'School Dist Code',
                                            'Urban-centric Locale [District] 2017-18':'Urban-centric Local',
                                           'Total Race/Ethnicity [District] 2017-18':'Total_student_Dist',
                                           'Free and Reduced Lunch Students [Public School] 2017-18':'Free_Reduced_Lunch_Students'})
schools_type
#df.rename(columns={"A": "a", "B": "c"})
All_school_type = pd.merge(schools, schools_type, how="left", on='School Dist Code')
All_school_type.head(5)
```

```python
All_school_type["Free_Reduced_Lunch_Students"] = pd.to_numeric(All_school_type["Free_Reduced_Lunch_Students"])
All_school_type['Reduced_Lunch_Students_%'] = ((All_school_type['Free_Reduced_Lunch_Students'] /
                                                All_school_type['Total_student_Dist'])*100)

All_school_type.head(5)
```

```python
Urban_centric = set(All_school_type['Urban-centric Local'])
Urban_centric
```

```python
All_school_type.loc[(All_school_type['Urban-centric Local'].str.contains("City")),'Urban-centric Local']='City'
All_school_type.loc[(All_school_type['Urban-centric Local'].str.contains("Suburb")),'Urban-centric Local']='Suburb'
All_school_type.loc[(All_school_type['Urban-centric Local'].str.contains("Town")),'Urban-centric Local']='Town'
All_school_type.loc[(All_school_type['Urban-centric Local'].str.contains("Rural")),'Urban-centric Local']='Rural'
#df.loc[(df.Event == 'Dance'),'Event']='Hip-Hop'
#df[df["A"].str.contains("Hello|Britain")]
All_school_type
```

```python
#school_type_group = All_school_type.groupby(['Test Score Bracket', 'Urban-centric Local'])['Dist Avg Score'].count()
#school_type_group = All_school_type.groupby(['Urban-centric Local']).count()#['Test Score Bracket'].count()
#school_type_group

town_df = All_school_type.loc[All_school_type["Urban-centric Local"] == "Town", :]
town_df
```

```python
av_df = All_school_type.loc[All_school_type["Test Score Bracket"] == "Average", :]
av_df = av_df.groupby(["Urban-centric Local"])["School Dist Code"].count()
av_df
```

```python
ba_df = All_school_type.loc[All_school_type["Test Score Bracket"] == "Below Average", :]
ba_df = ba_df.groupby(["Urban-centric Local"])["School Dist Code"].count()
ba_df
```

```python
#date = Southern_city_data['Date'].max()
x_Student_Free=All_school_type['Reduced_Lunch_Students_%']
y_Score=All_school_type['Dist Avg Score']
ratio_Score = All_school_type['Agency Name'].index
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_Student_Free, y_Score)
#print(f"the intercept :{round(intercept,2)}");
regress_values_Score = x_Student_Free * slope + intercept
line_Score = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
correlation_Score = round(rvalue,2)
plt.style.use('seaborn')
sur = plt.scatter(x_Student_Free, y_Score, s=100, c=ratio_Score, cmap="Blues_r", edgecolor="grey", linewidth=1, alpha=0.75);
plt.plot(x_Student_Free,regress_values_Score,"black");
plt.annotate(line_Score,(20,970),fontsize=14, color="black")
plt.title(f"student free Cities vs. Score SAT", fontsize = "12")
plt.xlabel(f"% Students by School in Free Lunch", size=14)
plt.ylabel(f"Combined Test Score", size=14)
#plt.tight_layout()
cbar= plt.colorbar()
cbar.set_label('Combined Test Score')
plt.savefig("../Image/scatter_chart_scores_SAT.png")
plt.show();
print(f"the intercept :{round(intercept,3)}");
print(f"correlation:{round(correlation_Score,3)}");
print(f"Lat linear regression model :{line_Score}");
print(f"===================================================")
print(f"There is a strong negative correlation between Student in program Free Luch and reduce price\
. the correlation value is {round(correlation_Score*100,2)}% in fact the greater the number of students\
in the free or reduced price lunch program, the combined SAT score decreases directly proportionally.")
```
