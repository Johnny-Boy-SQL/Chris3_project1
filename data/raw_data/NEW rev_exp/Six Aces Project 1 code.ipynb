```python
import os
import pandas as pd
```

```python
#convert SAS data to pandas dataframes for each of revenue & expenditure (re) data
re1415_df=pd.read_sas("raw_data/rev_exp/revexp_district_1415.sas7bdat",format='sas7bdat',iterator=False,chunksize=None,encoding="utf-8") #https://pandas.pydata.org/docs/reference/api/pandas.read_sas.html
state = re1415_df['STABBR']
gare1415_df = re1415_df.loc[(state == "GA")]

re1516_df=pd.read_sas("raw_data/rev_exp/revexp_district_1516.sas7bdat",format='sas7bdat',iterator=False,chunksize=None,encoding="utf-8") #https://pandas.pydata.org/docs/reference/api/pandas.read_sas.html
state = re1516_df['STABBR']
gare1516_df = re1516_df.loc[(state == "GA")]

re1617_df=pd.read_sas("raw_data/rev_exp/revexp_district_1617.sas7bdat",format='sas7bdat',iterator=False,chunksize=None,encoding="utf-8") #https://pandas.pydata.org/docs/reference/api/pandas.read_sas.html
state = re1617_df['STABBR']
gare1617_df = re1617_df.loc[(state == "GA")]

re1718_df=pd.read_sas("raw_data/rev_exp/revexp_district_1718.sas7bdat",format='sas7bdat',iterator=False,chunksize=None,encoding="utf-8") #https://pandas.pydata.org/docs/reference/api/pandas.read_sas.html
state = re1718_df['STABBR']
gare1718_df = re1718_df.loc[(state == "GA")]
```

```python
#Next, we merge the data sets on the LEAID variable, the unique variable for the districts across the years
gare1416=pd.merge(gare1415_df,gare1516_df,on="LEAID",suffixes=("_1415", "_1516"))
#gare1416.count()
gare1417=pd.merge(gare1416,gare1617_df,on="LEAID",suffixes=("_1416","_1617"))
#gare1417.count()
gare1418_df=pd.merge(gare1417,gare1718_df,on="LEAID",suffixes=("_1417","_1718"))
gare1418_df.head()
```

```python
import csv

gare1418_df.to_csv("rev_exp_bydistrict_2014-2018.csv")
```
